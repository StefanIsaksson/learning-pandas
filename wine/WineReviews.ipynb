```python
def rescale_points(row):
    row.points = (row.points - review_points_min)/2
    return row

reviews.apply(rescale_points, axis='columns')
```

```python
germany_score = reviews[reviews.country == 'Germany'].groupby(['points']).points.count()
```

```python
germany_score.plot.bar()
```

```python
france_score = reviews[reviews.country == 'France'].groupby(['points']).points.count()
```

```python
france_score.plot.bar()
```

```python
combined_country_score = pd.DataFrame(
    {'Germany': germany_score, 
     'France' : france_score}
)
```

```python
combined_country_score
```

## Stacked plots

```python
combined_country_score.plot.bar(stacked=True)
```

```python
combined_country_score_relative = combined_country_score.copy()
combined_country_score_relative.France = combined_country_score_relative.France / combined_country_score_relative.France.sum()
combined_country_score_relative.Germany = combined_country_score_relative.Germany / combined_country_score_relative.Germany.sum()
```

```python
combined_country_score_relative
```

```python
combined_country_score_relative.plot.bar(stacked=True)
```

```python
combined_country_score_relative.plot.area()
```
