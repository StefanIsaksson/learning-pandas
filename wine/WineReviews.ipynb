Map takes one set of values and "maps" them to another set of values.

```python
review_points_min = reviews.points.min()
```

```python
new_point_scale = reviews.points.map(lambda p: (p - review_points_min)/2)
```

```python
new_point_scale.describe()
```

Dataframe.apply calls a custom method on each row.

```python
def rescale_points(row):
    row.points = (row.points - review_points_min)/2
    return row

reviews.apply(rescale_points, axis='columns')
```
